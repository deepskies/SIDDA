```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Create a GridSpec with variable column widths to add spacing between groups
fig = plt.figure(figsize=(23, 7))

# Specify width_ratios directly in the GridSpec constructor
gs = gridspec.GridSpec(2, 9, width_ratios=[.3, .3, 0.1, .3, .3, 0.1, .3, .3, 0.1])

# Adjust wspace (width space) and hspace (height space) as needed
gs.update(wspace=0.0, hspace=0.3)

# First group (Left Square)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[1, 0])
ax3 = plt.subplot(gs[0, 1])
ax4 = plt.subplot(gs[1, 1])

# Second group (Middle Square)
ax5 = plt.subplot(gs[0, 3])
ax6 = plt.subplot(gs[1, 3])
ax7 = plt.subplot(gs[0, 4])
ax8 = plt.subplot(gs[1, 4])

# Third group (Right Square)
ax9 = plt.subplot(gs[0, 6])
ax10 = plt.subplot(gs[1, 6])
ax11 = plt.subplot(gs[0, 7])
ax12 = plt.subplot(gs[1, 7])

# Collect all axes for easy iteration
axs = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11, ax12]

# Display images and set titles
ax1.imshow(shapes[2])
ax1.set_title('Circle')
ax2.imshow(shapes_noisy[9])
ax2.set_title('Line')
ax3.imshow(shapes[4])
ax3.set_title('Rectangle')
ax4.imshow(shapes_noisy[1])
ax4.set_title('Circle')

ax5.imshow(blobs[2])
ax5.set_title('Star')
ax6.imshow(blobs_noisy[2])
ax6.set_title('Spiral')
ax7.imshow(blobs[3])
ax7.set_title('Elliptical')
ax8.imshow(blobs_noisy[6])
ax8.set_title('Star')

ax9.imshow(mnistm[4])
ax9.set_title('5')
ax10.imshow(mnistm_noisy[2])
ax10.set_title('1')
ax11.imshow(mnistm[5])
ax11.set_title('2')
ax12.imshow(mnistm_blur[11])
ax12.set_title('8')

# Set title font sizes and remove ticks
for ax in axs:
    ax.title.set_fontsize(18)
    ax.set_xticks([])
    ax.set_yticks([])

# Add group labels if needed
ax1.set_ylabel('Source \nDomain', fontsize=20, rotation=0, labelpad=50)
ax2.set_ylabel('Target \nDomain', fontsize=20, rotation=0, labelpad=50)

# Adjust layout
# plt.tight_layout()
plt.savefig('/Users/snehpandya/Projects/GCNN_DA/plots/ex_images.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
